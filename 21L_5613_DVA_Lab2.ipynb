```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import csv
```

```python
url = "https://www.dawnnews.tv/latest-news"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
data = []

for article in soup.find_all('article', class_='story'):
    story_title = article.find('h2', class_='story__title')
    if story_title:
        story_title = story_title.text.strip()
        print("Story Title:", story_title)
    else:
        story_title = ''

    story_link = article.find('a', class_='story__link')
    if story_link:
        story_link = story_link['href']
        print("Story Link:", story_link)
    else:
        story_link = ''

    media_item = article.find('div', class_='media__item')
    media_src = ""
    if media_item:
        media_img = media_item.find('img')
        if media_img and 'src' in media_img.attrs:
            media_src = media_img['src']
        print("Media Item:", media_src)

    story = article.find('div', class_='story__excerpt')
    if story:
        story = story.text.strip()
        print("Story Excerpt:", story)
    else:
        story = ''

    story_time = article.find('span', class_='story__time')
    if story_time:
        story_time = story_time.text.strip()
        print("Story Time:", story_time)
    else:
        story_time = ''

    print("\n")

    data.append([story_title, story_link, media_src, story, story_time])

csv_file = "21L-5613_Lab2.csv"

with open(csv_file, 'w', newline='', encoding='utf-8-sig') as csvfile:
    writer = csv.writer(csvfile)

    writer.writerow(["Story Title", "Story Link", "Media Item", "Story Excerpt", "Story Time"])

    for row in data:
        writer.writerow(row)

print("All the data saved to ", csv_file)
```
